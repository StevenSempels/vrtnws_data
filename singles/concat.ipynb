```python
import pandas as pd
```

```python
huis_15 = pd.read_csv('huishoudens_2015.csv', sep=';')
huis_16 = pd.read_csv('huishoudens_2016.csv', sep=';')
huis_17 = pd.read_csv('huishoudens_2017.csv', sep=';')
huis_18 = pd.read_csv('huishoudens_2018.csv', sep=';')
huis_19 = pd.read_csv('huishoudens_2019.csv', sep=';')
```

```python
huis_15.columns = huis_15.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('-', '')
huis_16.columns = huis_16.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('-', '')
huis_17.columns = huis_17.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('-', '')
huis_18.columns = huis_18.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('-', '')
huis_19.columns = huis_19.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('-', '')
```

```python
huis_15.info()
```

```python
huis_15 = huis_15[['refnis', '0.totaal_private_huishoudens', '1.alleenwonenden', '6.alleenstaande_ouders', '7.andere_types__huishoudens']]
huis_16 = huis_16[['refnis', '0.totaal_private_huishoudens', '1.alleenwonenden', '6.alleenstaande_ouders', '7.andere_types__huishoudens']]
huis_17 = huis_17[['refnis', '0.totaal_private_huishoudens', '1.alleenwonenden', '6.alleenstaande_ouders', '7.andere_types__huishoudens']]
huis_18 = huis_18[['refnis', '0.totaal_private_huishoudens', '1.alleenwonenden', '6.alleenstaande_ouders', '7.andere_types__huishoudens']]
huis_19 = huis_19[['woonplaats', '0.totaal_private_huishoudens', '1.alleenwonenden', '6.alleenstaande_ouders', '7.andere_types__huishoudens']]
```

```python
huis_15.rename(columns={'refnis':'woonplaats', '0.totaal_private_huishoudens':'totaal', '1.alleenwonenden':'alleenwonenden', '6.alleenstaande_ouders':'alleenstaande_ouders', '7.andere_types__huishoudens':'andere_huishoudens'}, inplace=True)
huis_16.rename(columns={'refnis':'woonplaats', '0.totaal_private_huishoudens':'totaal', '1.alleenwonenden':'alleenwonenden', '6.alleenstaande_ouders':'alleenstaande_ouders', '7.andere_types__huishoudens':'andere_huishoudens'}, inplace=True)
huis_17.rename(columns={'refnis':'woonplaats', '0.totaal_private_huishoudens':'totaal', '1.alleenwonenden':'alleenwonenden', '6.alleenstaande_ouders':'alleenstaande_ouders', '7.andere_types__huishoudens':'andere_huishoudens'}, inplace=True)
huis_18.rename(columns={'refnis':'woonplaats', '0.totaal_private_huishoudens':'totaal', '1.alleenwonenden':'alleenwonenden', '6.alleenstaande_ouders':'alleenstaande_ouders', '7.andere_types__huishoudens':'andere_huishoudens'}, inplace=True)
huis_19.rename(columns={'0.totaal_private_huishoudens':'totaal', '1.alleenwonenden':'alleenwonenden', '6.alleenstaande_ouders':'alleenstaande_ouders', '7.andere_types__huishoudens':'andere_huishoudens'}, inplace=True)
```

```python
huis_15.insert(1, 'jaar', 2015)
huis_16.insert(1, 'jaar', 2016)
huis_17.insert(1, 'jaar', 2017)
huis_18.insert(1, 'jaar', 2018)
huis_19.insert(1, 'jaar', 2019)
```

```python
huis_19.head()
```

```python
huishoudens = pd.concat([huis_15, huis_16, huis_17, huis_18, huis_19], sort=True)
```

```python
huishoudens = huishoudens[['woonplaats', 'jaar', 'totaal', 'alleenwonenden', 'alleenstaande_ouders', 'andere_huishoudens']]
```

```python
huishoudens.head()
```

```python
huishoudens['alleen_per_duizend'] = huishoudens.alleenwonenden * (1/(huishoudens.totaal/1000))
```

```python
huishoudens['alleen_per_duizend'] = huishoudens.alleen_per_duizend.round(0).astype(int)
```

```python
huishoudens.head()
```

```python
huishoudens['ouders_per_duizend'] = huishoudens.alleenstaande_ouders * (1/(huishoudens.totaal/1000))
```

```python
huishoudens['ouders_per_duizend'] = huishoudens.ouders_per_duizend.round(0).astype(int)
```

```python
huishoudens['anders_per_duizend'] = huishoudens.andere_huishoudens * (1/(huishoudens.totaal/1000))
```

```python
huishoudens['anders_per_duizend'] = huishoudens.anders_per_duizend.round(0).astype(int)
```

```python
huishoudens.head()
```

```python
huishoudens.to_csv('huishoudens_belgie.csv', sep=';', index=False, encoding='utf-8')
```

```python
bevolking = pd.read_csv('bevolking_geslacht_leeftijd20002019.csv')
```

```python
bevolking.columns = bevolking.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('-', '')
```

```python
bevolking.info()
```

```python
bevolking = bevolking[['gemeente', 'geslacht', 'leeftijdsklasse', 'bevolking_op_01_januari_2015', 'bevolking_op_01_januari_2016', 'bevolking_op_01_januari_2017', 'bevolking_op_01_januari_2018', 'bevolking_op_01_januari_2019']]
```

```python
bevolking.bevolking_op_01_januari_2015 = bevolking.bevolking_op_01_januari_2015.fillna(0)
bevolking.bevolking_op_01_januari_2016 = bevolking.bevolking_op_01_januari_2016.fillna(0)
bevolking.bevolking_op_01_januari_2017 = bevolking.bevolking_op_01_januari_2017.fillna(0)
bevolking.bevolking_op_01_januari_2018 = bevolking.bevolking_op_01_januari_2018.fillna(0)
bevolking.bevolking_op_01_januari_2019 = bevolking.bevolking_op_01_januari_2019.fillna(0)
```

```python
bevolking.bevolking_op_01_januari_2015 = bevolking.bevolking_op_01_januari_2015.astype(int)
bevolking.bevolking_op_01_januari_2016 = bevolking.bevolking_op_01_januari_2016.astype(int)
bevolking.bevolking_op_01_januari_2017 = bevolking.bevolking_op_01_januari_2017.astype(int)
bevolking.bevolking_op_01_januari_2018 = bevolking.bevolking_op_01_januari_2018.astype(int)
bevolking.bevolking_op_01_januari_2019 = bevolking.bevolking_op_01_januari_2019.astype(int)
```

```python
bevolking.head()
```

```python
bevolking = bevolking.melt(id_vars=['gemeente', 'geslacht', 'leeftijdsklasse'], value_vars=['bevolking_op_01_januari_2015', 'bevolking_op_01_januari_2016', 'bevolking_op_01_januari_2017', 'bevolking_op_01_januari_2018', 'bevolking_op_01_januari_2019'], var_name='jaar', value_name='aantal')
```

```python
bevolking.jaar = bevolking.jaar.str.replace('bevolking_op_01_januari_', '')
```

```python
bevolking.head()
```

```python
leeftijden = ['Van 15 tot 19 jaar', 'Van 20 tot 24 jaar', 'Van 25 tot 29 jaar']
jonge_mensen = bevolking[bevolking.leeftijdsklasse.isin(leeftijden) & bevolking.gemeente.notna()]
jonge_mensen = jonge_mensen.groupby(['gemeente', 'jaar']).agg({'aantal' : 'sum'})
```

```python
jonge_mensen = jonge_mensen.reset_index()
```

```python
totaal = bevolking[bevolking.leeftijdsklasse.isna()]
```

```python
totaal = totaal[['gemeente', 'jaar', 'aantal']]
```

```python
merged = jonge_mensen.merge(totaal, on=['gemeente', 'jaar'], how='left')
```

```python
merged.rename(columns={'aantal_x':'jong', 'aantal_y':'totaal'}, inplace=True)
```

```python
merged['per_duizend'] = merged.jong * (1/(merged.totaal/1000))
```

```python
merged.per_duizend = merged.per_duizend.round(0).astype(int)
```

```python
merged.head()
```

```python
merged.jaar = merged.jaar.astype(int)
```

```python
new = huishoudens.merge(merged, left_on=['woonplaats', 'jaar'], right_on=['gemeente', 'jaar'], how='left')
```

```python
new = new.dropna()
```

```python
new.jong = new.jong.astype(int)
new.totaal_y = new.totaal_y.astype(int)
new.per_duizend = new.per_duizend.astype(int)
```

```python
new.head()
```

```python
import seaborn as sb
```

```python
pearson = new.corr(method='pearson')
```

```python
sb.heatmap(pearson, 
            xticklabels=pearson.columns,
            yticklabels=pearson.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```

Conclusie op basis van hierboven: geen sterke correlatie tussen de rato van jonge mensen (15-30 jaar oud) en rato van alleenwonenden/alleenstaande ouders/overige huishoudens in een gemeente. Wel een middelmatige correlatie tussen de rato van alleenwonenden en de rato van 'overige huishoudens' (zoals vrienden die samenwonen).
